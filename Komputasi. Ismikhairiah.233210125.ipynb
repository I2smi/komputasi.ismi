```python

import numpy as np
import matplotlib.pyplot as plt

# Data seismik (contoh data sederhana)
data_seismik = np.array([
    [10, 12, 15, 18, 20],
    [11, 13, 16, 19, 21],
    [12, 14, 17, 20, 22],
    [13, 15, 18, 21, 23],
    [14, 16, 19, 22, 24]
])

# Parameter reservoir
ketebalan_reservoir = 20  # meter
porositas = 0.2  # fraksi
faktor_recovery = 0.5  # fraksi

# Menentukan batas reservoir (contoh sederhana)
batas_atas = 2
batas_bawah = 3
batas_kiri = 1
batas_kanan = 3

# Menghitung volume reservoir
volume_reservoir = (batas_kanan - batas_kiri) * (batas_bawah - batas_atas) * ketebalan_reservoir

# Menghitung volume minyak dan gas
volume_minyak = volume_reservoir * porositas * faktor_recovery
volume_gas = volume_reservoir * porositas * (1 - faktor_recovery)

# Menampilkan hasil
print("Volume Reservoir:", volume_reservoir, "meter kubik")
print("Volume Minyak:", volume_minyak, "meter kubik")
print("Volume Gas:", volume_gas, "meter kubik")

# Visualisasi data seismik
plt.imshow(data_seismik, aspect="auto", cmap="gray")
plt.xlabel("Sampel")
plt.ylabel("Waktu (s)")
plt.title("Data Seismik")

# Menampilkan batas reservoir
plt.axvline(x=batas_kiri, color="green", linestyle="--")
plt.axvline(x=batas_kanan, color="green", linestyle="--")
plt.axhline(y=batas_atas, color="green", linestyle="--")
plt.axhline(y=batas_bawah, color="green", linestyle="--")

plt.show()
```
